```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('G:\Kaggle_compitation\Linear Regression\Dataset\day.csv')
df.head()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(18,12))
sns.boxplot(x="season", y="cnt", data=df, ax=ax1)
sns.boxplot(x="weathersit", y="cnt", data=df, ax=ax2)
sns.boxplot(x="holiday", y="cnt", data=df, ax=ax3)
sns.boxplot(x="workingday", y="cnt", data=df, ax=ax4)
fig.show()
```


Above we have plotted all the categorial variables distribution with respect to the count/sales variable.

we do not see any outliers in any of our variable.
we can also conclude that we can have more number of riders in Summer and Fall compared to other seasons.
we can expect more sales when we have a clear weather.
we do not see if any difference in sales when it comes to working or non-working day.



```python

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18,6))
sns.scatterplot(x="temp", y="cnt", data=df, ax=ax1)
sns.scatterplot(x="hum", y="cnt", data=df, ax=ax2)
sns.scatterplot(x="windspeed", y="cnt", data=df, ax=ax3)
fig.show()
```

Checking if there is any Linear relationship between our continous variables and our target variable.

we see some linear relation between count and the temp variable.
for the other 2 variables we cannot say the same.(no pattern)



```python

# let us look at the correation matrix of the data
import numpy as np
corr = df[['temp','atemp','hum','windspeed','cnt']].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))
# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,square=True, linewidths=.5, annot=True)
f.show()
```
